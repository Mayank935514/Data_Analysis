```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Data_Set/AB_NYC_2019.csv")
```

```python
df.head(2)
```

## Histogram

```python
sns.histplot(df["price"])
```

```python
sns.histplot(df[df["price"]<500]["price"])
```
