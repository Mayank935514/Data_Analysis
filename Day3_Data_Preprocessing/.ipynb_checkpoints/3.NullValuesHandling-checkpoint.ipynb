```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
ls
```

```python
df = pd.read_csv("Data.csv")
```

## Null Values Handling--> Numeric

```python
df
```

## Approch-1. Removing the rows

```python
df.dropna()
```

## Approach-2. Imputers

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

imputer.fit(df.iloc[:,1:3].values)

df.iloc[:,1:3] = imputer.transform(df.iloc[:,1:3].values)
```

```python
df
```

## Null Values Handling--> Categorical

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:] = imputer.transform(df.iloc[:,:].values)
```

```python
df
```

## Null Values Handling on Google Playstore Dataset

```python
data = pd.read_csv("Google_Playstore/googleplaystore.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
impute = SimpleImputer(missing_values = np.nan , strategy = 'most_frequent')
impute.fit(data.iloc[ : , 2:3 ].values)
data.iloc[ : , 2:3 ] = impute.transform(data.iloc[ : , 2:3 ].values)
```
