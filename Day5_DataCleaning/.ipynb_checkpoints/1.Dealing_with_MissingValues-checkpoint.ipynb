## Filling the Missing Values--> Imputation

```python
df3 = pd.read_csv("Data_Set/AB_NYC_2019.csv")
```

```python
df3.shape
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
mean_Value = df3['reviews_per_month'].mean()
```

```python
df3.fillna({'reviews_per_month': mean_Value}, inplace=True)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
df3[df3['number_of_reviews'] == 0].head()
```

## Filling Categorical Data--> Use mode values

```python
df3.nunique()
```

```python
df3.fillna({'last_review': df3['last_review'].value_counts().index[0]}, inplace = True)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

## Filling Categorical Data--> Use new category

```python
df = pd.read_csv('Data_Set/AB_NYC_2019.csv')
df4=df.copy()
df4.fillna({'last_review': 'Not Reviewed'}, inplace = True)
```

```python
df4.sample(10)
```
